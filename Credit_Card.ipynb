```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("CC GENERAL.csv")
```

```python
df.head()
```

```python
inertia_scores=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df[['PURCHASES','PAYMENTS']])
    inertia_scores.append(kmeans.inertia_)
plt.plot(range(1,11),inertia_scores)
```

```python
clustering=KMeans(3)
clustering.fit(df[['PURCHASES','PAYMENTS']])
df['PURCHASE_PAYMENT_CLUSTER']=clustering.labels_
```

```python
centers=pd.DataFrame(clustering.cluster_centers_)
centers.columns=['x','y']
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df,x='PURCHASES',y='PAYMENTS',hue='PURCHASE_PAYMENT_CLUSTER',palette='tab10')
plt.scatter(x=centers['x'],y=centers['y'],s=100,color='black',marker="+")
```
